Converting hour entries from float to integer

```python
df_train['Hour'] = pd.to_datetime(df_train['Hour'],format='%H:%M')
df_train['Hour'] = df_train['Hour'].dt.hour
df_test['Hour'] = pd.to_datetime(df_test['Hour'],format='%H:%M')
df_test['Hour'] = df_test['Hour'].dt.hour
df_train.dtypes
```

Converting Date Column entry from object to pandas date-time integer

```python
df_train['Date'] = pd.to_datetime(df_train['Date'])
df_test['Date'] = pd.to_datetime(df_test['Date'])
df_train.dtypes
```

Seperating the Year, month, day to seperate coloumns;

```python
df_train['Year'] = df_train['Date'].dt.year
df_train['Month'] = df_train['Date'].dt.month
df_train['Day'] = df_train['Date'].dt.day_of_week
df_test['Year'] = df_test['Date'].dt.year
df_test['Month'] = df_test['Date'].dt.month
df_test['Day'] = df_test['Date'].dt.day_of_week
df_train.head(7)
```

```python
df_ids=df_test['id']
```

Dropping the coloumns ID, Date as they are not required now.

```python
df_train = df_train.drop(columns=['Date','id'])
df_test = df_test.drop(columns=['Date','id'])
df_train.head(5)
```

Encoding the coloumns of seasons, holiday and working day.

```python
seasons_train=df_train.Seasons.value_counts().to_dict()

i=1
for key in seasons_train:
  seasons_train[key]=i
  i=i+1

print(seasons_train)
df_train['Seasons']=df_train['Seasons'].map(seasons_train)


holiday_train=df_train.Holiday.value_counts().to_dict()
i=0
for key in holiday_train:
  holiday_train[key]=i
  i=i+1

print(holiday_train)
df_train['Holiday']=df_train['Holiday'].map(holiday_train)



df_train.rename(columns={'Working Day': 'Working_Day'}, inplace="True")
working_day=df_train.Working_Day.value_counts().to_dict()

for key in holiday_train:
  if(holiday_train[key]==1):
    working_day[key]=1
  else:
    working_day[key]=0


print(working_day)
df_train['Working_Day']=df_train['Working_Day'].map(working_day)




# **********************************************************************************************

df_test['Seasons']=df_test['Seasons'].map(seasons_train)

df_test['Holiday']=df_test['Holiday'].map(holiday_train)

df_test.rename(columns={'Working Day': 'Working_Day'}, inplace="True")
df_test['Working_Day']=df_test['Working_Day'].map(working_day)
```
